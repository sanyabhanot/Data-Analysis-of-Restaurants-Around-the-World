# L3.T1: Restaurant Reviews

**Analyze the text reviews to identify the most common positive and negative keywords.**

```python
data['Rating text'].value_counts()
```

```python
!pip install textblob
```

```python
from textblob import TextBlob
from collections import Counter

rating_texts = data['Rating text'].dropna()
positive_keywords = []
negative_keywords = []

for text in rating_texts:
    analysis = TextBlob(text)

    if analysis.sentiment.polarity > 0:
        positive_keywords.extend([text.lower()])
    elif analysis.sentiment.polarity < 0:
        negative_keywords.extend([text.lower()])
```

```python
positive_keyword_counts = Counter(positive_keywords)
negative_keyword_counts = Counter(negative_keywords)

print("Most common positive keywords: ", *positive_keyword_counts.most_common())
print("Most common negative keywords: ", *negative_keyword_counts.most_common())
```

*The most frequently occurring positive keyword is "**good**," while the most prevalent negative keyword is "**average**"*

```python
plt.figure(figsize=(5, 10))
plt.pie(positive_keyword_counts.values(), labels=positive_keyword_counts.keys(), autopct='%1.1f%%')
plt.title('Positive Sentiment Distribution')
plt.show()
```

```python
plt.figure(figsize=(5, 10))
plt.pie(negative_keyword_counts.values(), labels=negative_keyword_counts.keys(), autopct='%1.1f%%')
plt.title('Negative Sentiment Distribution')
plt.show()
```

```python
all_reviews = []

for text in rating_texts:
    all_reviews.extend([text.lower()])

review_counts = Counter(all_reviews)

plt.figure(figsize=(5, 10))
plt.pie(review_counts.values(), labels=review_counts.keys(), autopct='%1.1f%%', startangle=140)
plt.title('All Reviews Distribution')
plt.show()
```

**Calculate the average length of reviews and explore if there is a relationship between review length and rating.**

```python
data['Review length'] = data['Rating text'].apply(lambda x: len(str(x)))
average_review_length = data['Review length'].mean()

print(f"Average Review Length: {average_review_length:.2f} characters")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Review length'], data['Aggregate rating'], alpha=0.1)
plt.title('Relationship between Review Length and Rating')
plt.xlabel('Review Length (characters)')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
sentiment_mapping = {
    'Good': 'Positive',
    'Very Good': 'Positive',
    'Excellent': 'Positive',
    'Poor': 'Negative',
    'Not rated': 'Neutral',
    'Average': 'Neutral',
}

data['Sentiment'] = data['Rating text'].map(sentiment_mapping)
from scipy.stats import pearsonr
sentiment_mapping = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1,
}
data['Sentiment_numeric'] = data['Sentiment'].map(sentiment_mapping)

correlation_coefficient, _ = pearsonr(data['Review length'], data['Sentiment_numeric'])

print(f"Correlation Coefficient: {correlation_coefficient}")
```

*A correlation coefficient of **-0.3255** suggests a **moderately negative** correlation between review length and sentiment*

```python
data['Review Length'] = data['Rating text'].str.len()
data['Review Length']
average_review_length = data.groupby('Sentiment')['Review Length'].mean()
average_review_length.plot(kind='bar')
plt.xlabel('Sentiment')
plt.ylabel('Average Review Length')
plt.title('Relationship Between Review Length and Sentiment')
plt.xticks(rotation=45)
plt.show()

print("Average Review Length by Sentiment:")
print(average_review_length)
```

Based on these average review lengths, it appears that **negative** reviews are the **shortest** and tend to be very concise. **Neutral** reviews are **slightly longer**, while **positive** reviews fall in between but are still **relatively short**. This suggests that customers who write negative reviews often express their sentiment in very **brief** statements, while those writing neutral and positive reviews may include slightly **more detail** in their feedback.
*However, it's important to note that these are average values and individual reviews can vary widely in length.*

**Determine if higher-priced restaurants are more likely to offer these services.**

```python
grouped_data = data.groupby('Price range').agg(
    Online_delivery_percentage=('Has Online delivery', lambda x: (x == 'Yes').mean() * 100),
    Table_booking_percentage=('Has Table booking', lambda x: (x == 'Yes').mean() * 100),
)

# Plot
ax = grouped_data.plot(kind='bar', figsize=(8, 6))
plt.xlabel('Price Range')
plt.ylabel('Percentage')
plt.title('Percentage of Restaurants Offering Services by Price Range')
plt.xticks(rotation=0)
plt.legend(title='Service')
plt.show()
```

```python
# Calculate the average price range for restaurants offering online delivery and table booking
avg_price_range_with_delivery = data[data['Has Online delivery'] == 'Yes']['Price range'].mean()
avg_price_range_with_table_booking = data[data['Has Table booking'] == 'Yes']['Price range'].mean()

print("Average Price Range for Restaurants Offering Online Delivery:", avg_price_range_with_delivery)
print("Average Price Range for Restaurants Offering Table Booking:", avg_price_range_with_table_booking)
```

```python
# Create a contingency table to analyze the relationship
contingency_table = pd.crosstab(index=data['Price range'], columns=[data['Has Online delivery'], data['Has Table booking']], margins=True)

# Print the contingency table
print(contingency_table)
```

```python
data['Has Table booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})
data['Has Online delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})

high_priced_offers = data[(data['Price range'] >= 3) & (data['Has Table booking']) & (data['Has Online delivery'])]

result = high_priced_offers[['Has Table booking', 'Has Online delivery']].mean()

if (result['Has Table booking'] == 1) and (result['Has Online delivery'] == 1):
    print("Yes, higher-priced restaurants are offering both services(Table Booking and Online Delivery).")
else:
    print("No, higher-priced restaurants are offering both services(Table Booking and Online Delivery).")
```
